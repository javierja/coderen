```python
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
# Define the ticker symbol for Pfizer
ticker = ['PFE']

# Seteo las fechas desde y hasta (meses donde comienza la pandemia)
start_date = "2020-02-01"
end_date = "2020-04-01"

# Fetch the historical data using yfinance
data = yf.download(ticker, start=start_date, end=end_date)
data
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data['Close'])
plt.title('Pfizer Stock Price')
plt.xlabel('Date')
plt.ylabel('Closing Price ($)')
plt.grid(True)
plt.show()
#En el grÃ¡fico puede verse un fuerte caÃ­da de ambas empresas hacia marzo de 2020
```

```python
# Introduzco las fechas desde que se comienza a comercializar la vacuna del COVID-19, durante un periodo de 2 aÃ±os (11-20 y 11-22)
covid_start_date = "2020-11-01"
covid_end_date = "2022-11-01"

# Obtengo la informacion historia durante el periodo de comercializacion de la vacuna
covid_data = yf.download(ticker, start=covid_start_date, end=covid_end_date)
# covid_data['Close']=covid_data.index
# covid_data=covid_data.reset_index(drop=True)
covid_data
```

```python
# Obtengo un grÃ¡fico con el precio de cierre* durante el periodo previamente seteado (de comercializacion de la vacuna)
# *El precio de cierre, tambiÃ©n conocido como la cotizaciÃ³n de cierre, es el precio de un activo en el mercado de inversiones al final del dÃ­a de operaciones. 
plt.figure(figsize=(12, 6))
plt.plot(covid_data['Close'])
plt.title('Pfizer - Precio Periodo de venta de la vacuna - COVID-19')
plt.xlabel('Date')
plt.ylabel('Closing Price ($)')
plt.grid(True)
plt.show()
#En el grafico se ve como la curva va en aumento y llega casi a duplicar los valores que antes se analizaron, marzo de 2020
```

```python
#Sobre el data frame obtenido voy a realizar algunos cambios:
#primero vere las 5 pimeras filas
covid_data.head()

#luego vere cuantos columnas posee el DF
covid_data.columns

#luego vere cuantos datos (filas) posee el DF
len(covid_data)

#Como solo me interesan algunos datos dejare 3 columnas y la mitad de las filas (250) como muestreo

dconvert = covid_data.drop(columns=['High', 'Low', 'Adj Close'])

dconvert = dconvert.sample(n = 250)

#Ahora me aseguro que los datos no vengan duplicados

dconvert.duplicated()
dconvert = dconvert.drop_duplicates()
#Reemplazo un valor por otro (485000000 por 48.50)

dconvert = dconvert.replace(39.730000,39.9999999)

#Finalmente renombro los nombres de las dos primeras columnas
dconvert = dconvert.rename(columns = {"Open": "OpenPrice", "Close": "ClosePrice"})

dconvert
```
